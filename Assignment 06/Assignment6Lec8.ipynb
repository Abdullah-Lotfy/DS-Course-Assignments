```python
import os
import random
import string
```

```python
def ran_nam_gen(size=6, letters=string.ascii_uppercase):
    return ''.join(random.choice(letters) for x in range(size))
```

```python
ran_nam_gen()
```

```python
file_mix = os.listdir('file_mix')
```

```python
for i in file_mix:
    file_name = i.split('.')[0]
    file_extension = i.split('.')[-1]
    if file_extension == 'mp3' or file_extension == 'jpg':
        continue
    else:
        os.rename(f'file_mix/{i}',f'file_mix/{ran_nam_gen()}.{file_extension}')
```
